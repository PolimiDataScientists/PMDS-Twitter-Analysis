```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
sentiments = pd.read_csv("../data/sentiments.csv")
sentiments = sentiments[["day","negPercentage","netPercentage","posPercentage"]]
```

```python
covidData = pd.read_csv("../data/covidData.csv")
covidData = covidData[["date","deathIncrease","positiveIncrease"]]
covidData.head()
```

```python
start_date = '2020-04-15'
end_date = '2020-05-30'
mask = (covidData['date'] >= start_date) & (covidData['date'] <= end_date)
covidData = covidData.loc[mask]
covidData = covidData.sort_values(by=['date'])
```

```python
covidData.sort_values(by=['date'])
```

```python
sentiments.plot('day', "negPercentage")
covidData.plot("date", "deathIncrease")
covidData.plot("date", "positiveIncrease")
```
